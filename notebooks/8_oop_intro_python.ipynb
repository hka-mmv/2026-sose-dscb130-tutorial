Python ist eine objektorientierte Programmiersprache. Fast alles in Python ist ein Objekt, zusammen mit seinen Eigenschaften (Attributen) und Methoden. Objektorientierte Programmierung (OOP) hilft uns dabei, unseren Code logisch zu strukturieren, reale Konzepte nachzubauen und Code wiederverwendbar zu machen.

## Klassen und Objekte
Eine Klasse (Class) ist wie ein Konstrukteur oder ein "Bauplan" für die Erstellung von Objekten. 
Um beim Bauplan zu bleiben: Die Klasse ist die technische Zeichnung für ein Haus. Das Objekt ist das tatsächlich gebaute Haus. Aus einem Bauplan können wir beliebig viele Häuser (Objekte) bauen.

### Eine Klasse erstellen
Um eine Klasse zu erstellen, verwendt man das Schlüsselwort `class`:

```python
class MyClass:
  x = 5
```

### Ein Objekt erstellen
Jetzt kann man die Klasse `MyClass` verwenden, um Objekte zu erstellen:

```python
p1 = MyClass()
print(p1.x)
```

## Die `__init__()` Funktion
Das obige Beispiel mit `x = 5` ist in der Praxis selten nützlich. Um Klassen wirklich zu verstehen, muss man sich die integrierte `__init__()`-Funktion (den Konstruktor) ansehen.

Alle Klassen haben eine Funktion namens `__init__()`, die immer ausgeführt wird, wenn die Klasse initiiert (also ein neues Objekt erstellt) wird. Man nutzt diese Funktion, um den Eigenschaften des Objekts Startwerte zuzuweisen.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

# Wir erstellen ein Person-Objekt mit Namen und Alter
p1 = Person("John", 36)

print(p1.name)
print(p1.age)
print(p1)
```

*Wichtig:* Die `__init__()` Funktion wird jedes Mal **automatisch** aufgerufen, wenn ein neues Objekt aus einer Klasse erstellt wird.

## Die `__str__()` Funktion
Die `__str__()` Funktion steuert, was zurückgegeben werden soll, wenn das Klassenobjekt als String dargestellt wird (z. B. wenn man `print(p1)` aufruft). Ohne diese Funktion wird nur der Speicherort des Objekts ausgegeben.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name} ist {self.age} Jahre alt."

p1 = Person("John", 36)
print(p1)
```

## Objektmethoden
Objekte können auch Methoden enthalten. Methoden sind Funktionen, die in einer Klasse definiert sind und zu den Objekten gehören. Sie definieren, was ein Objekt "tun" kann.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hallo, mein Name ist " + self.name)

p1 = Person("John", 36)
p1.myfunc()
```

## Der `self` Parameter
Der `self`-Parameter ist eine Referenz auf die aktuelle Instanz der Klasse und wird verwendet, um auf Variablen oder Methoden zuzugreifen, die zur Klasse gehören.

Er **muss** nicht zwingend `self` heißen. Aber er muss **immer der erste Parameter** in jeder Funktion einer Klasse sein! Das Wort `self` ist eine extrem starke Konvention in Python, an die sich alle Entwickler halten.

## Eigenschaften ändern oder löschen
Eigenschaften von Objekten können wie normale Variablen überschrieben werden.

```python
# Das Alter von p1 auf 40 ändern
p1.age = 40

# Man kann Eigenschaften auch komplett löschen (eher selten in der Praxis)
del p1.age

# Oder das ganze Objekt löschen
del p1
```

## Vererbung (Inheritance)
Die Vererbung ermöglicht es, eine Klasse zu definieren, die alle Methoden und Eigenschaften einer anderen Klasse übernimmt. Das ist extrem praktisch, um Code nicht doppelt schreiben zu müssen (DRY-Prinzip: Don't Repeat Yourself).


Man unterscheidet zwischen zwei Begriffen:
* **Elternklasse (Parent Class / Base Class):** Die Klasse, von der geerbt wird.
* **Kindklasse (Child Class / Derived Class):** Die Klasse, die von einer anderen Klasse erbt.

### Eine Elternklasse erstellen

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def print_info(self):
    print(f"Name: {self.name}, Alter: {self.age}")

p1 = Person("John", 36)
p1.print_info()
```

### Eine Kindklasse erstellen
Man möchte nun eine Klasse namens `Student` erstellen. Ein Student ist logischerweise auch eine Person. Er sollte also einen Namen und ein Alter haben. Anstatt diese Eigenschaften neu zu programmieren, soll `Student` von `Person` erben.

Dazu schreibt man den Namen der Elternklasse in Klammern hinter den Namen der Kindklasse. Wenn man der Kindklasse vorerst keine neuen Funktionen hinzufügen wollen, nutzt man das Schlüsselwort `pass`.

```python
class Student(Person):
  pass

# Wir können das Student-Objekt nun genau wie ein Person-Objekt nutzen
s1 = Student("Mike", 20)
s1.print_info()
print(s1)
```

## Die `__init__()` Funktion in der Kindklasse
Bisher hat die `Student`-Klasse die `__init__()`-Funktion der Elternklasse einfach übernommen. Was passiert aber, wenn amn dem Studenten eine *eigene* `__init__()`-Funktion gibt?

Sobald an eine `__init__()`-Funktion in der Kindklasse definiert, **überschreibt** sie die der Elternklasse. Das Kind erbt die `__init__()`-Funktion der Eltern dann *nicht* mehr!

Damit das Kind trotzdem noch die Eigenschaften der Elternklasse initialisieren kann, müssen wir die Funktion der Elternklasse explizit aufrufen. Dafür gibt es die `super()` Funktion.

```python
class Student(Person):
  def __init__(self, name, age, graduation_year):
    # super() ruft automatisch die Methoden der Elternklasse (Person) auf
    super().__init__(name, age) 
    
    # Jetzt fügen wir eine ganz neue Eigenschaft hinzu, die nur Studenten haben
    self.graduation_year = graduation_year

s1 = Student("Mike", 20, 2026)
print(f"{s1.name} macht seinen Abschluss im Jahr {s1.graduation_year}.")
```

## Eigene Methoden zur Kindklasse hinzufügen

```python
class Student(Person):
  def __init__(self, name, age, graduation_year):
    super().__init__(name, age)
    self.graduation_year = graduation_year

  def welcome(self):
    print(f"Willkommen {self.name}! Dein Abschlussjahrgang ist {self.graduation_year}.")

s1 = Student("Mike", 20, 2026)

# s1 kann weiterhin print_info() von Person nutzen
s1.print_info()

# und zusätzlich seine eigene Methode welcome()!
s1.welcome()
```

## Polymorphismus (Vielgestaltigkeit)
Das Wort "Polymorphismus" bedeutet "viele Formen". In der Programmierung bezieht es sich auf Methoden, Funktionen oder Operatoren mit demselben Namen, die auf verschiedenen Objekten oder Datentypen ausgeführt werden können.

Ein klassisches, bereits bekanntes Beispiel in Python ist die integrierte Funktion `len()`. Sie kann verwendet werden, um die Länge von ganz verschiedenen Datentypen herauszufinden:

```python
# Die Länge eines Strings (Zeichenkette)
x = "Hello World"
print(len(x))

# Die Länge eines Tuples (oder einer Liste)
mytuple = ("apple", "banana", "cherry")
print(len(mytuple))
```

### Polymorphismus in Klassen
Polymorphismus wird oft in Klassen verwendet, wenn verschiedene Klassen denselben Methodennamen haben. 

Angenommen, es gibt drei verschiedene Klassen: `Car` (Auto), `Boat` (Boot) und `Plane` (Flugzeug). Alle drei haben eine Methode namens `move()`.

```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fahren!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Segeln!")

class Plane:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fliegen!")

car1 = Car("Ford", "Mustang")       
boat1 = Boat("Ibiza", "Touring 20") 
plane1 = Plane("Boeing", "747")     

# Da alle Objekte die Methode move() besitzen, kann man sie in einer Schleife aufrufen, 
# ohne den genauen Klassentyp überprüfen zu müssen:
for x in (car1, boat1, plane1):
  x.move()
```

### Polymorphismus und Vererbung
Wie im vorherigen Kapitel zur Vererbung behandelt, können Kindklassen Methoden von Elternklassen erben. Kombiniert man dies mit Polymorphismus, kann eine Methode der Elternklasse in der Kindklasse überschrieben werden, behält aber denselben Namen ("Method Overriding").

Es wird nun eine Basisklasse `Vehicle` (Fahrzeug) erstellt. Die Klassen `Car`, `Boat` und `Plane` erben von dieser Basisklasse.

```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Fahren!")

class Car(Vehicle):
  # Das Auto nutzt einfach die Standard-Methode der Elternklasse
  pass

class Boat(Vehicle):
  # Das Boot überschreibt die move() Methode
  def move(self):
    print("Segeln!")

class Plane(Vehicle):
  # Das Flugzeug überschreibt die move() Methode ebenfalls
  def move(self):
    print("Fliegen!")

car1 = Car("Ford", "Mustang")       
boat1 = Boat("Ibiza", "Touring 20") 
plane1 = Plane("Boeing", "747")     

for x in (car1, boat1, plane1):
  print(x.brand)
  x.move()
```

In diesem Beispiel erbt `Car` die Methode `move()` direkt von `Vehicle` und nutzt sie ohne Änderung (gibt also "Fortbewegen!" aus). `Boat` und `Plane` erben ebenfalls, überschreiben die Methode jedoch mit ihrem eigenen, spezifischen Verhalten.

## Cheatsheet: Polymorphismus
| Begriff | Beschreibung |
| :--- | :--- |
| `Polymorphismus` | Die Fähigkeit verschiedener Objekte, auf denselben Methodenaufruf jeweils spezifisch zu reagieren. |
| `Method Overriding` | Das Überschreiben einer geerbten Methode in der Kindklasse (ein häufiger Anwendungsfall von Polymorphismus in der Vererbung). |
