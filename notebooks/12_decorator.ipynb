Ein Dekorator (Decorator) ist ein mächtiges Konzept in Python, das es ermöglicht, das Verhalten einer Funktion oder Methode zu erweitern oder zu verändern, ohne den eigentlichen Quellcode dieser Funktion anzupassen. 

Man kann sich einen Dekorator wie eine Geschenkverpackung vorstellen: Das eigentliche Geschenk (die Funktion) bleibt unverändert, aber es bekommt eine zusätzliche Hülle (den Dekorator), die das Erscheinungsbild oder die Funktionsweise nach außen hin anpasst.

### Vorwissen: Funktionen als Objekte
Um Dekoratoren zu verstehen, muss man wissen, dass Funktionen in Python sogenannte *First-Class Objects* sind. Das bedeutet, Funktionen können wie ganz normale Variablen behandelt, an andere Funktionen übergeben oder von diesen zurückgegeben werden.

```python
def gruessen(name):
    return f"Hallo {name}!"

def verarbeite_funktion(andere_funktion, name):
    # Hier wird die übergebene Funktion ausgeführt
    ergebnis = andere_funktion(name)
    print(f"Die übergebene Funktion ergab: {ergebnis}")

# Wir übergeben die Funktion 'gruessen' OHNE Klammern (da sie nicht sofort ausgeführt werden soll)
verarbeite_funktion(gruessen, "Alice")
gruessen("Alice")  # Direkter Aufruf der Funktion mit Klammern
```

### Dekoratoren mit Parametern (`*args` und `**kwargs`)
Wenn die zu dekorierende Funktion Parameter akzeptiert, muss auch die innere `wrapper`-Funktion diese Parameter aufnehmen und weiterleiten. Um den Dekorator universell für jede Art von Funktion nutzbar zu machen, verwendet man `*args` und `**kwargs`.

### Einen eigenen Dekorator schreiben
Ein Dekorator ist im Grunde eine Funktion, die eine andere Funktion als Parameter annimmt. Innerhalb des Dekorators wird eine innere Funktion (oft `wrapper` genannt) definiert. Diese Wrapper-Funktion führt den zusätzlichen Code aus und ruft dann die ursprüngliche Funktion auf. Am Ende gibt der Dekorator die Wrapper-Funktion zurück.

Hier wird ein einfacher Dekorator geschrieben, der vor und nach dem Aufruf einer Funktion Text ausgibt:

```python
def mein_erster_dekorator(funktion):
    # Die innere Wrapper-Funktion
    def wrapper():
        print("Etwas passiert VOR dem Aufruf der Funktion.")
        funktion() # Aufruf der eigentlichen Funktion
        print("Etwas passiert NACH dem Aufruf der Funktion.\n")
    
    # Der Dekorator gibt die Wrapper-Funktion zurück
    return wrapper

def hallo_sagen():
    print("Hallo! Ich bin die Kernfunktion.")

# Manuelle Anwendung des Dekorators:
dekoriertes_hallo = mein_erster_dekorator(hallo_sagen)
dekoriertes_hallo()
```

### Die `@`-Syntax verwenden (Syntactic Sugar)
Die manuelle Zuweisung wie im obigen Beispiel (`dekoriertes_hallo = ...`) ist oft unübersichtlich. Python bietet dafür eine sehr elegante Abkürzung: das `@`-Symbol. 

Wird das `@`-Symbol gefolgt vom Namen des Dekorators direkt über die Definition einer Funktion geschrieben, wendet Python den Dekorator automatisch an.

```python
# Wir nutzen den gleichen Dekorator von oben
def mein_erster_dekorator(funktion):
    def wrapper():
        print("Etwas passiert VOR dem Aufruf der Funktion.")
        funktion()
        print("Etwas passiert NACH dem Aufruf der Funktion.\n")
    return wrapper

# Die @-Syntax wendet den Dekorator automatisch an
@mein_erster_dekorator
def hallo_sagen():
    print("Hallo! Ich bin die Kernfunktion.")

# Jetzt reicht ein normaler Funktionsaufruf
hallo_sagen()
```

```python
def log_aufruf(funktion):
    # *args und **kwargs fangen alle beliebigen Parameter ab
    def wrapper(*args, **kwargs):
        print(f"Die Funktion '{funktion.__name__}' wurde aufgerufen.")
        # Parameter werden an die Kernfunktion weitergereicht
        ergebnis = funktion(*args, **kwargs)
        return ergebnis
    return wrapper

@log_aufruf
def addiere(a, b):
    return a + b

@log_aufruf
def begruessung(name):
    return f"Moin {name}"

print(addiere(5, 3))
print(begruessung("Bob"))
```

## Cheatsheet: Dekoratoren
| Syntax / Konzept | Beschreibung |
| :--- | :--- |
| `@dekorator_name` | Wird über eine Funktion geschrieben, um das Verhalten dieser Funktion zu verändern/erweitern. |
| `Wrapper` | Die innere Funktion eines Dekorators, die den erweiterten Code sowie den Aufruf der Originalfunktion enthält. |
| `First-Class Objects` | Das Prinzip, dass Funktionen wie Variablen behandelt, übergeben und zurückgegeben werden können. |
| `*args, **kwargs` | Werden im Wrapper genutzt, damit der Dekorator mit Funktionen umgehen kann, die eine beliebige Anzahl von Parametern haben. |
