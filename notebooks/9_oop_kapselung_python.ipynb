## Kapselung (Encapsulation)
Kapselung ist ein weiteres Kernkonzept der objektorientierten Programmierung. Es beschreibt den Vorgang, Daten (Attribute) und die dazugehörigen Methoden innerhalb einer Klasse zu bündeln und den direkten Zugriff von außen auf diese Daten einzuschränken. 

Dies dient dem Schutz der Daten vor unbeabsichtigten Änderungen. Man kann sich das wie eine Kapsel in der Pharmazie vorstellen: Der Inhalt ist sicher im Inneren verstaut, und man interagiert nur mit der äußeren Hülle.

### Zugriffsmodifizierer in Python
Anders als Programmiersprachen wie Java oder C++ kennt Python keine strengen Zugriffsmodifizierer. In Python basiert der Schutz von Attributen stark auf **Konventionen** bei der Namensgebung:

* **Public (Öffentlich):** Der Standard. Attribute und Methoden können von überall aufgerufen werden (z. B. `self.name`).
* **Protected (Geschützt):** Wird durch einen einfachen Unterstrich signalisiert (z. B. `self._age`). Dies ist ein Hinweis an andere Programmierer: "Bitte greife nicht direkt auf dieses Attribut zu, es ist nur für den internen Gebrauch gedacht." Es ist jedoch technisch nicht blockiert.
* **Private (Privat):** Wird durch einen doppelten Unterstrich signalisiert (z. B. `self.__password`). Python verändert den Namen dieses Attributs intern ("Name Mangling"), sodass ein direkter Zugriff von außen extrem erschwert wird.

```python
class Bankkonto:
  def __init__(self, besitzer, startguthaben):
    self.besitzer = besitzer         # Public
    self._bankname = "Uni-Bank"      # Protected (Konvention)
    self.__guthaben = startguthaben  # Private

konto = Bankkonto("Alice", 1000)

# Public: Problemlos möglich
print(konto.besitzer) 

# Protected: Technisch möglich, aber man sollte es laut Konvention vermeiden!
print(konto._bankname)

# Private: Führt zu einem Fehler (AttributeError), wenn das Kommentarzeichen entfernt wird!
# print(konto.__guthaben)
```

### Getter und Setter Methoden
Wenn Attribute privat (`__`) gemacht werden, stellt sich die Frage, wie das Guthaben abgefragt oder geändert werden kann. Dafür werden sogenannte "Getter"- und "Setter"-Methoden definiert. Sie bilden eine kontrollierte Schnittstelle nach außen. 

So kann beispielsweise beim Abheben von Geld geprüft werden, ob das Guthaben überhaupt ausreicht, bevor der Wert intern geändert wird.

```python
class Bankkonto:
  def __init__(self, besitzer, startguthaben):
    self.besitzer = besitzer
    self.__guthaben = startguthaben

  # Getter-Methode: Gibt den Wert sicher zurück
  def get_guthaben(self):
    return self.__guthaben

  # Setter-Methode: Kontrolliert die Änderung des Wertes
  def abheben(self, betrag):
    if betrag > 0 and betrag <= self.__guthaben:
      self.__guthaben -= betrag
      print(f"{betrag}€ erfolgreich abgehoben.")
    else:
      print("Fehler: Ungültiger Betrag oder nicht ausreichendes Guthaben.")

konto = Bankkonto("Alice", 1000)

# Kontrollierter Zugriff über die Methoden:
print(f"Aktuelles Guthaben: {konto.get_guthaben()}€")

konto.abheben(300)
print(f"Neues Guthaben: {konto.get_guthaben()}€")

konto.abheben(2000) # Löst die Fehlerprüfung im Setter aus
```

## Cheatsheet: Kapselung
| Präfix | Bedeutung | Beschreibung |
| :--- | :--- | :--- |
| `(keines)` | Public | Das Attribut ist von überall zugänglich und veränderbar. |
| `_` | Protected | Konvention: Sollte nicht von außen angefasst werden. Technisch aber möglich. |
| `__` | Private | Schwer zugänglich von außen. Verhindert unbeabsichtigtes Überschreiben. |
| `Getter` | Methode | Eine Funktion, um den Wert eines privaten Attributs sicher auszulesen. |
| `Setter` | Methode | Eine Funktion, um den Wert eines privaten Attributs kontrolliert zu ändern. |

## Innere Klassen (Inner Classes)
In Python ist es möglich, eine Klasse innerhalb einer anderen Klasse zu definieren. Dies wird als "innere Klasse" (Inner Class) oder verschachtelte Klasse bezeichnet.

Innere Klassen werden hauptsächlich verwendet, um logisch verwandte Klassen zu gruppieren. Wenn eine Klasse ausschließlich für eine andere Klasse von Bedeutung ist, kann sie in dieser "äußeren" Klasse gekapselt werden. Dies sorgt für einen besser strukturierten und aufgeräumten Code.

### Eine innere Klasse erstellen
Um eine innere Klasse zu erstellen, wird das Schlüsselwort `class` einfach innerhalb des Blocks einer anderen Klasse verwendet.

```python
class OuterClass:
  def __init__(self):
    print("Äußere Klasse wird erstellt.")

  # Definition der inneren Klasse
  class InnerClass:
    def __init__(self):
      print("Innere Klasse wird erstellt.")
```

### Objekte einer inneren Klasse instanziieren
Es gibt zwei gängige Ansätze, um ein Objekt einer inneren Klasse zu erstellen und zu nutzen. 

**Weg 1: Instanziierung innerhalb der äußeren Klasse**
Oft wird das Objekt der inneren Klasse direkt in der `__init__()`-Funktion der äußeren Klasse erzeugt. So ist sichergestellt, dass die innere Struktur automatisch existiert, sobald das äußere Objekt erstellt wird.

```python
class Person:
  def __init__(self, name):
    self.name = name
    # Das Gehirn-Objekt wird automatisch als Teil der Person erstellt
    self.brain = self.Brain()

  class Brain:
    def __init__(self):
      self.status = "denkt nach..."

    def think(self):
      print("Das Gehirn " + self.status)

p1 = Person("Alice")

# Zugriff auf die Methode der inneren Klasse über das äußere Objekt
print(p1.name)
p1.brain.think()
```

**Weg 2: Direkte Instanziierung von außen**
Alternativ kann man eine innere Klasse auch direkt von außen aufrufen, ohne vorher ein Objekt der äußeren Klasse komplett initialisieren zu müssen. Dafür muss der Name der äußeren Klasse dem der inneren vorangestellt werden.

```python
class Haus:
  class Zimmer:
    def __init__(self, funktion):
      self.funktion = funktion

    def zeige_funktion(self):
      print(f"Dieses Zimmer ist das {self.funktion}.")

# Direkte Erstellung eines Zimmer-Objekts über die Klasse Haus
mein_zimmer = Haus.Zimmer("Schlafzimmer")
mein_zimmer.zeige_funktion()
```

## Cheatsheet: Innere Klassen
| Begriff / Syntax | Beschreibung |
| :--- | :--- |
| `Innere Klasse` | Eine Klasse, die innerhalb einer anderen Klasse definiert ist, um zusammenhängende Konzepte zu bündeln. |
| `self.inner = self.Inner()` | Gängiger Weg, um die innere Klasse direkt bei der Erstellung der äußeren Klasse zu instanziieren. |
| `Outer.Inner()` | Syntax, um ein Objekt der inneren Klasse direkt von außen zu erzeugen. |

## Dunder-Methoden (Magic Methods)
In Python gibt es spezielle Methoden, die mit doppelten Unterstrichen beginnen und enden (z.B. `__init__` oder `__str__`). Diese werden oft als **Dunder**-Methoden (von "Double UNDERscore") oder "Magic Methods" bezeichnet. 

Sie sind "magisch", weil sie meist nicht direkt aufgerufen werden (wie `objekt.__str__()`), sondern von Python automatisch im Hintergrund verwendet werden, wenn bestimmte integrierte Funktionen oder Operatoren auf ein Objekt angewendet werden.

Wir haben bereits zwei der wichtigsten kennengelernt:
* `__init__(self)`: Wird beim Erstellen eines Objekts aufgerufen.
* `__str__(self)`: Wird aufgerufen, wenn ein Objekt in einen String umgewandelt wird (z. B. bei `print()`).

### Weitere wichtige Dunder-Methoden
Mit Dunder-Methoden können wir definieren, wie sich unsere eigenen Objekte bei Standard-Operationen verhalten sollen.

**1. `__len__(self)` - Längenbestimmung**
Definiert, was passiert, wenn die Funktion `len()` auf ein Objekt angewendet wird.

```python
class Bibliothek:
  def __init__(self):
    self.buecher = ["Python Grundlagen", "Data Science", "KI für Anfänger"]

  def __len__(self):
    # Gibt die Anzahl der Bücher in der Liste zurück
    return len(self.buecher)

meine_bib = Bibliothek()
print(f"Anzahl der Bücher: {len(meine_bib)}")
```

**2. Operator Overloading (z. B. `__add__` für `+`)**
Wir können festlegen, was passiert, wenn wir zwei Objekte mit dem `+` Operator addieren. Dies nennt man Operator Overloading.

```python
class Vektor:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __add__(self, anderer_vektor):
    # Addiert die X- und Y-Werte beider Vektoren
    neues_x = self.x + anderer_vektor.x
    neues_y = self.y + anderer_vektor.y
    return Vektor(neues_x, neues_y)

  def __str__(self):
    return f"Vektor({self.x}, {self.y})"

v1 = Vektor(2, 4)
v2 = Vektor(3, 1)

# Hier wird im Hintergrund v1.__add__(v2) aufgerufen!
ergebnis = v1 + v2 
print(ergebnis)
```

## Cheatsheet: Häufige Dunder-Methoden
| Methode | Wird aufgerufen durch... | Beschreibung |
| :--- | :--- | :--- |
| `__init__(self)` | `Objekt()` | Initialisiert das Objekt. |
| `__str__(self)` | `print(obj)` oder `str(obj)` | Gibt eine lesbare Textdarstellung zurück. |
| `__len__(self)` | `len(obj)` | Gibt die Länge/Größe des Objekts zurück. |
| `__add__(self, other)` | `obj1 + obj2` | Definiert die Addition. |
| `__eq__(self, other)` | `obj1 == obj2` | Definiert den Vergleich auf Gleichheit. |
