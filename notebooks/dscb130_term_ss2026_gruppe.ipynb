# Aufgabe 1 - Startmenü und TUI-Struktur
1. Startmenü:  
Erstellen eines Hauptfensters mit der Spieleauswahl (TicTacToe, Connect-4 und Schiffeversenken). Bei Auswahl wird ein neues Menü mit der Spieleauswahl angezeigt.
2. Einbau eines Spielerregisters:  
Auswahlmöglichkeit zum erstellen von Spielern. Bei Auswahl wird ein neues Menü mit dem Spielerregister geöffnet.
3. Einbau von Spielerstatistik:
Auswahlmöglichkeit zum Öffnen des Spielerstatistikmenüs.
4. Stoppfunktion im Startmenü:
Die Anwendung muss durch eine "Beende"-Funktion gestoppt werden können, ohne das Statistik oder das Spielerregister beschädigt werden. (Ausnahme im laufenden Spielbetrieb, z.B. bei Spielcrash)
5. Interaktion mit dem Startmenü:  
Durch Eingabe einer Zahl (Auswahlmöglichkeit) soll die Aktion ausgewählt werden.
6. Einfache Fehlerbehandlung:  
Umgang mit ungültigen Eingaben. Anzeigen einer Fehlermeldung oder Rücksetzen der Eingabe. Das Startmenü darf nicht beendet werden (Crash), außer die Stoppfunktion wird verwendet.

## Erwartete Ergebnisse
- Ein funktionsfähiges Hauptfenster mit eingebauten Auwahlmöglichkeiten:
(1) Spieleauswahl, (2) Spielerregister, (3) Spielerstatistik, (4) Beenden
- Strukturierter, kommentierter Code als Ausgangspunkt für die nächsten Aufgaben

# Aufgabe 2 - Spielerregister (Untermenü)
Zugriff auf Untermenü nur über das Startmenü
1. Auswahlmöglichkeit zur Erstellung eines neuen Spielers. Dabei wird eine unique ID erstellt, die dem Spieler eindeutig zuweisbar ist. Über eine Eingabe kann der Spieler einen eigenen Namen angeben (mehrfachbelegung muss möglich sein). Außerdem auch das jeweilige Lieblingsspiel (Auswahl aus TicTacToe, 4-Gewinnt und Schiffeversenken).
2. Das Spielerregister wird in einem .json-Format permanent gespeichert und kann gegen Manipulation geschützt werden (Siehe Aufgabe 4)
2. Spielerregister muss auf Abruf alle Spielernamen mit ID auflisten.
3. Spieler müssen gelöscht werden können. Dabei werden alle zu dem Spieler gehörende Spielstände ebenfalls gelöscht.

## Erwartete Ergebnisse
- Funktionsfähiges Spielermanagement und Aufruf aus permanenter .json-Datei

# Aufgabe 3 - Spielerstatistik (Untermenü)
1. Erweiterung des .json-Speichers:  
Spielstand der aktiven Spieler (ID) muss zum Ende einer Partie gespeichert werden (Art des Spiels, Gegner (PvP, PvCOM), Anzahl der Züge, Sieg/Niederlage)
2. Bestenliste:  
Bestenliste aller registrierten Spieler, filterbar auf Top 3, Top 10 oder All; für jedes Spiel.
3. Spielerstatistik:  
Individuelle Spielerstatistik: Bestes Spiel mit Sieg/Niederlagerate, filterbar nach Gegnertyp (PvP, PvCOM), Durchschnittliche Spieler
4. Computerstatistik muss nicht gespeichert werden.

## Erwartete Ergebnisse

# Aufgabe 5 - Spiellogik
Implementierung der Kernmechanik.
1. Modulares Spielmenü zur Auswahl des Spiels:
Auswahl aus TicTacToe, 4-Gewinnt und Schiffeversenken.
2. Spielerauswahl basierend auf ID:
- ID wird eingegeben um die Spieler (nacheinander) zu bestimmen
- Einbau einer COM Auswahlfunktion (auch wenn Aufgabe 6 nicht gelöst, stattdessen Fehler werfen und neu auswählen)
- Zufälliges Bestimmen des Beginners im Spiel
3. Funktionale Spiellogik entwickeln (Engine):
Implementieren Sie die grafische Darstellung (Konsolen-/Terminalanwendung) und die Regeln für das jeweilige Spiel.
- Stateless Logic: Die Spiellogik (z. B. Gewinnprüfung, Validierung von Zügen) muss in „reinen“ Funktionen (Pure Functions) gekapselt sein. Diese erhalten das Spielfeld als Input und geben das Ergebnis (Sieg/Niederlage/Validität) zurück.
- Unveränderlichkeit: Das Spielfeld darf nicht „in-place“ manipuliert werden. Jeder Zug erzeugt eine neue Kopie des Spielfelds mit dem aktualisierten Status.
- Abbruchbedingungen: Implementieren Sie eine saubere Behandlung von Spielabbrüchen, damit Teilstatistiken nicht korrumpiert werden.
- Eingaben über Konsolenzeile

## Erwartete Ergebnisse
- Vollständige Spielzyklen: Alle drei Spiele sind vollumfänglich spielbar.
- Modus-Flexibilität: Die Anwendung kann nahtlos zwischen menschlichen Spielern und COM unterscheiden.
- Saubere Schnittstellen: Die Spiellogik ist strikt von der TUI-Logik getrennt (Separation of Concerns).

## Hinweise und Ressourcen
- Modus-Abfrage: Nach der Wahl eines Spiels (z. B. TicTacToe) muss bestimmt werden, ob gegen einen lokalen Mitspieler (aus dem Register) oder den COM-Solver (aus Aufgabe 5) angetreten wird.
- Funktionale Injektion: Nutzen Sie das Prinzip der Dependency Injection. Übergeben Sie dem Spiel-Loop eine „Zug-Funktion“. Bei PvP ist dies die Funktion zur Benutzereingabe, bei PvCOM die Solver-Funktion.
- Architektur-Tipp: Betrachten Sie ein Spiel als einen State-Reducer: (Alter_Zustand, Aktion) -> Neuer_Zustand
- Listen-Verarbeitung: Nutzen Sie Pythons all() und any() Funktionen für die Gewinnprüfung (z. B. um Reihen oder Spalten effizient zu prüfen).

# Aufgabe 6 - Funktionale Solver & PvCOM
Ziel ist es, Algorithmen zu entwickeln, die basierend auf dem aktuellen Spielfeld einen (nicht zwingend optimalen, aber nicht random basierten) Spielzug berechnen und als COM durchführen. Wählen Sie mindestens 2 Spiele aus (TicTacToe, 4-Gewinnt, Schiffeversenken).
1. Zustandsbasierte Entscheidungen:
Der Solver darf keine globalen Variablen verändern. Er erhält den aktuellen Spielstatus als Read-Only-Datenstruktur und gibt lediglich die Koordinate des nächsten Zuges zurück.
2. Integration in den Spielprozess:
Der Wechsel zwischen menschlichem Spieler und Computer muss nahtlos erfolgen.

## Erwartete Ergebnisse
- Ein funktionierender COM-Modus für die gewählten Spiele
- Statistik-Koppelung: Ergebnisse gegen den COM müssen in der Statistik erkennbar sein

## Hinweise und Ressourcen
- Für einfache Spiele (TicTacToe) bietet sich ein Minimax-Algorithmus an.
- [Minimax Wikipedia](https://de.wikipedia.org/wiki/Minimax-Algorithmus)
- Funktionaler Tipp: Nutzen Sie map oder List Comprehensions, um alle möglichen Züge zu bewerten, anstatt klassische for-Schleifen mit Zwischenzuständen zu verwenden.
- Integrationstipp: "Higher-Order Functions" --> Versuchen Sie, den Spielablauf so zu abstrahieren, dass die Funktion z.B. get_next_move austauschbar ist – je nachdem, ob gerade ein Mensch (input) oder der Solver (calculate_move) am Zug ist.
