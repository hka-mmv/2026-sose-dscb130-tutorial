## Intro to Git

### Git Setup
Nach der Installation von Git, muss einmalig die Git Umgebung (Environment) aufgesetzt werden.

#### Die Identität
Der erste Schritt nach der Installation von Git ist das setzen der eigenen Identität. Diese wird beim erstellen von Commits als Signatur verwendet.
```console
$ git config --global user.name "John Doe"
$ git config --global user.email johndoe@example.com
```

#### Der "default"-Branch Name
Wird ein neues Git Repository ("Repo"/"repo" folgend) erstellt, wird der erste branch üblicherweise "master" genannt. Was ein branch oder repo ist, wird später erklärt. Um diesen Standardnamen zu ändern:
```console
$ git config --global init.defaultBranch NAME
```
Hierbei wird "NAME" mit dem neuen Standardnamen ersetzt.

#### Troubleshooting
Git verfügt über eine ausführliche Dokumentation zu allen Funktionen.
```console
$ git help <verb>
$ git <verb> --help
$ man git-<verb>
```

## Lokaler Workflow
Bevor wir Daten zu GitHub hochladen, müssen wir verstehen, wie Git Änderungen lokal auf Rechner verwaltet. Git unterscheidet dabei zwischen drei Bereichen:
+ Working Directory: Der Projektordner, in dem Dateien bearbeitet werden (z.B. Jupyter Notebook).
+ Staging Area (Index): Ein Zwischenspeicher. Sammeln von Änderungen, die für den nächsten Snaphot bereit sind.
+ Local Repository: Speicher von Git der endgültigen Versionen (Commits) in einer Datenbank.

### Ein neues Repository initialisieren
Um einen Ordner unter Versionskontrolle zu stellen, nutzt man den init Befehl.
```console
$ git init
```
Dies erstellt einen versteckten Ordner .git, in dem die gesamte Historie gespeichert wird.

### Workflow
In Repositories arbeiten wir iterativ.
1. Status prüfen
Überprüfen, welche Dateien geändert wurden oder noch nicht von Git verfolgt werden:
```console
$ git status
```
2. Änderungen vermerken (staging)
Wenn eine Datei fertig bearbeitet wurde, wird sie der staging area hinzugefügt. File dient hierbei als placeholder für den Dateinamen. Um alle Änderungen gleichzeitig hinzuzufügen, verwendet man den 2. Befehl.
```console
$ git add file.py
$ git add . .
```
3. Änderungen speichern (commit)
Ein Commit ist ein Speicherpunkt (vergleichbar mit einem Videospiel). Er hält den aktuellen Zustand von staging fest. Zum Commit wird eine Nachricht mit kurzer Erklärung der Änderungen hinzugefügt, hier mit "message" abgekürzt.
```console
$ git commit -m "message"
```
4. Historie ansehen (check)
Um zu sehen, wer wann etwas geändert hat, oder um zu einem alten Punkt zurückzukehren, nutzt man:
```console
$ git log --oneline
```

Häufig zu Commiten eignet sich eher als seltene, "große" Commits zu machen. Durch kleinere, häufige Commits ist das Nachvollziehen von Fehlern und der Rollback einfacher.

## Remote Workflow
Nachdem wir den lokalen Workflow beherrschen, verbinden wir unser lokales Repository mit GitHub, um Code zu sichern und mit anderen zu teilen.

### Ein Remote Repository hinzufügen
Um ein lokales Repo mit einem leeren Repository auf GitHub zu verknüpfen, muss die URL des Remote-Ziels definiert werden:
```console
$ git remote add origin https://github.com/NUTZERNAME/REPO-NAME.git
```
"origin" ist hierbei der Standardname für den entfernten Server.

### Zusammenarbeit (Cloning)
Möchte man an einem Projekt mitarbeiten, das bereits auf GitHub existiert, kopiert man das gesamte Repository auf den eigenen Rechner:
```console
$ git clone https://github.com/NUTZERNAME/REPO-NAME.git
```
Dies erstellt automatisch einen Ordner mit dem Projektnamen, initialisiert Git und setzt das "origin" Remote-Ziel.

### Daten übertragen (Push & Pull)
Um lokale Änderungen auf GitHub hochzuladen oder Änderungen von dort zu empfangen, nutzen wir:
1. Änderungen hochladen
```console
$ $ git push -u origin main
```
Der Parameter -u verknüpft den lokalen Branch dauerhaft mit dem Remote-Branch.

2. Änderungen herunterladen
Falls Teammitglieder Änderungen hochgeladen haben, holen wir diese mit pull ab:
```console
$ $ git pull origin main
```
Dies aktualisiert das lokale Working Directory auf den neuesten Stand von GitHub.

### Verzweigung (Branching)
Branches erlauben es uns, an neuen Features zu arbeiten, ohne den funktionierenden Hauptcode ("main") zu gefährden. Ein Branch ist wie eine Abzweigung in der Zeitlinie des Projekts.
1. Neuen Branch erstellen
```console
$ git branch feature-name
```
2. Zu einem Branch wechseln
```console
$ git checkout feature-name
```
Alternativ erstellt und wechselt der Befehl git checkout -b NAME sofort in einen neuen Branch.

3. Zusammenführen (Merge)
Wenn die Arbeit im Feature-Branch fertig ist, wechseln wir zurück zum Hauptbranch und führen die Änderungen zusammen:
```console
$ git checkout main
$ git merge feature-name
```
Nach dem Merge sind die Änderungen aus feature-name im main Branch enthalten.

## Ignorieren von Dateien (.gitignore)
Nicht alle Dateien in einem Projektordner sollten von Git versioniert werden (z. B. temporäre Dateien, Log-Files oder Passwörter). Hierfür nutzt man eine .gitignore Datei.
### Funktionsweise
In der Textdatei .gitignore im Hauptverzeichnis des Repos werden Dateinamen oder Muster gelistet, die Git ignorieren soll.  
Beispielinhalt der Datei:
```plaintext
# Ignoriere alle .log Dateien
*.log

# Ignoriere den Ordner mit temporären Daten
/temp/

# Ignoriere Systemdateien von MacOS
.DS_Store
```
Dateien, die bereits committet wurden, werden nicht nachträglich ignoriert. Diese müssen erst aus dem Index entfernt werden.

## Umgang mit Mergekonflikten
Ein Konflikt entsteht, wenn zwei Personen die gleiche Zeile in derselben Datei unterschiedlich geändert haben und Git nicht automatisch entscheiden kann, welche Version korrekt ist.
### Konfliktlösung
Wenn Git einen "Merge Conflict" meldet, markiert es die betroffenen Stellen direkt in der Datei:
```plaintext
<<<<<<< HEAD
Meine lokale Änderung (Version A)
=======
Die Änderung aus dem anderen Branch (Version B)
>>>>>>> feature-name
```
Schritte zur Behebung:
1. Datei öffnen: Den Text zwischen den Markern (<<<<<<<, =======, >>>>>>>) manuell bearbeiten und entscheiden, was bleiben soll.
2. Marker entfernen: Alle von Git eingefügten Trennlinien löschen.
3. Änderung markieren: Die bereinigte Datei wieder zur Staging Area hinzufügen:
```console
$ git add file.py
```
4. Abschluss: Den Merge durch einen finalen Commit beenden:
```console
$ git commit -m "Fix merge conflict"
```

## Cheatsheet
| Befehl | Beschreibung |
| :--- | :--- |
| `git pull` | Lädt Änderungen vom Server und führt sie lokal zusammen. |
| `git push` | Lädt lokale Commits auf den Server hoch. |
| `git branch` | Zeigt alle lokalen Branches an. |
| `git checkout -b <name>` | Erstellt einen neuen Branch und wechselt sofort hinein. |
| `git merge <name>` | Führt den angegebenen Branch in den aktuellen Branch zusammen. |
