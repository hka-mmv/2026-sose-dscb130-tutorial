Wenn Programme größer und komplexer werden, ist es unpraktisch und fehleranfällig, den gesamten Code in einer einzigen Datei zu behalten. Modularisierung ist der Prozess, bei dem Code in kleinere, besser handhabbare und wiederverwendbare Teile aufgeteilt wird.

### Was ist ein Modul?
Ein Modul in Python ist schlichtweg eine Datei mit der Endung `.py`, die Python-Code (Variablen, Funktionen, Klassen) enthält. 

### Module importieren
Um den Code aus einem Modul in einer anderen Datei (oder in diesem Notebook) zu verwenden, nutzt man das Schlüsselwort `import`. Python bringt bereits eine große "Standardbibliothek" an fertigen Modulen mit, die direkt verwendet werden können.

````python
# Importieren des integrierten Mathematik-Moduls
import math

# Zugriff auf Funktionen und Variablen des Moduls erfolgt über den Modulnamen und einen Punkt
wurzel = math.sqrt(16)
print(f"Die Wurzel aus 16 ist {wurzel}")

# Zugriff auf eine Konstante aus dem Modul
print(f"Der Wert von Pi ist ca. {math.pi}")
````

## Verschiedene Import-Methoden
Es gibt mehrere Wege, Code aus einem Modul zu importieren. Welche man wählt, hängt davon ab, wie viel man aus dem Modul benötigt und wie übersichtlich der Code bleiben soll.

````python
# 1. Nur spezifische Funktionen importieren (erspart das ständige Voranstellen des Modulnamens)
from math import sqrt, pi

# Nun kann 'sqrt' direkt ohne 'math.' aufgerufen werden
print(sqrt(25)) 

# 2. Einem Modul einen kürzeren Alias geben (sehr häufig bei großen Bibliotheken)
import datetime as dt

heute = dt.date.today()
print(f"Heute ist der {heute}")
````

## Pakete (Packages)
Wenn ein Projekt aus sehr vielen Modulen besteht, können diese in Verzeichnissen (Ordnern) strukturiert werden. Ein Paket ist im Grunde ein Dateiordner, der mehrere Python-Module enthält.

Damit Python einen Ordner zweifelsfrei als Paket erkennt, platziert man traditionell eine Datei namens `__init__.py` (diese darf auch komplett leer sein) in diesem Ordner.

Ein beispielhafter Projektaufbau könnte so aussehen:
```plaintext
mein_projekt/
│
├── hauptprogramm.py
└── netzwerk/                 # Das ist ein Paket (Ordner)
    ├── __init__.py
    ├── wlan.py               # Das ist ein Modul
    └── bluetooth.py          # Das ist ein weiteres Modul

Soll nun im hauptprogramm.py eine Funktion aus der wlan.py genutzt werden, erfolgt der Import über die Punkt-Notation:

````python
from netzwerk import wlan
# oder
from netzwerk.wlan import verbinde_mit_router
````

## Externe Pakete (pip)
Neben der Standardbibliothek gibt es hunderttausende externe Pakete, die von der Community entwickelt wurden (z.B. NumPy für Mathematik, Pandas für Datenanalyse).
Diese werden nicht mit Python mitgeliefert, sondern müssen einmalig über das Terminal mit dem Paketmanager pip installiert werden:

$ pip install numpy

Nach der Installation können sie genau wie Standardmodule mit import im Code verwendet werden.

## Cheatsheet: Module und Pakete
| Befehl / Begriff | Beschreibung |
| :--- | :--- |
| `Modul` | Eine einzelne `.py` Datei, die Python-Code enthält. |
| `Paket` | Ein Ordner, der mehrere Module und eine `__init__.py` Datei enthält. |
| `import x` | Importiert das gesamte Modul `x`. Zugriff auf Inhalte via `x.funktion()`. |
| `from x import y` | Importiert nur das Element `y` aus dem Modul `x`. Zugriff direkt via `y()`. |
| `import x as y` | Importiert das Modul `x` unter dem Wunschnamen/Alias `y`. |
| `pip` | Der Standard-Paketmanager zur Installation externer Python-Bibliotheken über das Terminal. |
