Python ist von Natur aus eine "dynamisch typisierte" Programmiersprache. Das bedeutet, der Datentyp einer Variablen wird zur Laufzeit automatisch erkannt und kann sich theoretisch jederzeit ändern. 

Um Code – besonders in großen Projekten – lesbarer und sicherer zu machen, wurden "Type Hints" (Typ-Hinweise) eingeführt. 

**Wichtig:** Type Hints verändern das Verhalten des Codes zur Laufzeit *nicht* (Python erzwingt die Typen nicht). Sie dienen primär:
1. Der besseren Lesbarkeit für andere Programmierer.
2. Der Autovervollständigung in Entwicklungsumgebungen (IDEs).
3. Der Fehlersuche durch statische Analyse-Tools (wie `mypy`), bevor der Code überhaupt ausgeführt wird.

### Grundlegende Syntax
Man gibt den Datentyp einer Variablen mit einem Doppelpunkt `:` an. Den Rückgabetyp einer Funktion definiert man mit einem Pfeil `->`.

```python
# 'name' soll ein String sein, die Funktion gibt einen String zurück
def begruessung(name: str) -> str:
    return f"Hallo {name}!"

print(begruessung("Informatik-Erstis"))
```

### Type Hinting in Klassen
In der objektorientierten Programmierung sind Type Hints besonders nützlich, um auf einen Blick zu dokumentieren, welche Datentypen eine Klasse erwartet und welche Art von Objekten ihre Methoden zurückgeben.

```python
class Buch:
  # Type Hints für die Parameter des Konstruktors
  def __init__(self, titel: str, seiten: int):
    # Auch den Klassenattributen können Typen zugewiesen werden
    self.titel: str = titel
    self.seiten: int = seiten

  # Die Methode erwartet einen Integer und gibt einen Boolean (Wahr/Falsch) zurück
  def ist_gelesen(self, gelesene_seiten: int) -> bool:
    return gelesene_seiten >= self.seiten

mein_buch = Buch("Python Crashkurs", 350)
print(mein_buch.ist_gelesen(400))
```

### Komplexe Datentypen und eigene Klassen als Typen
Häufig möchte man angeben, dass eine Variable eine Liste von bestimmten Objekten enthalten soll, oder dass ein Rückgabewert auch "nichts" (`None`) sein darf.

* **Listen/Dictionaries:** Werden mit eckigen Klammern präzisiert, z. B. `list[int]` oder `dict[str, int]`.
* **Optionale Werte:** Wenn ein Wert entweder ein bestimmter Typ oder `None` sein kann, nutzt man den `|` Operator (z. B. `str | None`).
* **Eigene Klassen:** Klassen können selbst als Type Hint verwendet werden.

```python
class Buchregal:
  def __init__(self):
    # Eine Liste, die ausschließlich Objekte der Klasse 'Buch' enthalten soll
    self.buecher: list[Buch] = []

  # 'buch' muss ein Objekt der Klasse 'Buch' sein. 'None' als Rückgabe (Funktion gibt nichts zurück)
  def buch_hinzufuegen(self, buch: Buch) -> None:
    self.buecher.append(buch)

  # Gibt entweder ein 'Buch'-Objekt zurück oder 'None', falls es nicht gefunden wurde
  def suche_buch(self, titel: str) -> Buch | None:
    for b in self.buecher:
      if b.titel == titel:
        return b
    return None

regal = Buchregal()
regal.buch_hinzufuegen(mein_buch)

gefundenes_buch = regal.suche_buch("Python")
if gefundenes_buch:
    print(f"Gefunden: {gefundenes_buch.titel}")
```

## Cheatsheet: Type Hinting
| Syntax | Beschreibung |
| :--- | :--- |
| `variable: typ` | Definiert den erwarteten Datentyp einer Variablen oder eines Parameters. |
| `-> typ:` | Definiert den erwarteten Rückgabetyp einer Funktion/Methode. |
| `-> None:` | Signalisiert, dass eine Funktion/Methode keinen Wert zurückgibt (`return` ohne Wert). |
| `list[Typ]` | Eine Liste, die nur Elemente des angegebenen Typs enthält. |
| `Typ1 \| Typ2` | Der Wert kann entweder Typ1 oder Typ2 sein (z. B. `int \| None`). |
